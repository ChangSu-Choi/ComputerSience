```python
import numpy as np
```

```python
parent_A = 1258
parent_B = 3939
left_child_A = 1177
left_child_B = 1745
right_child_A = 81
right_child_B = 2194

def gini(A, B):
    x = np.array([A, B])
    return 1-((x/x.sum())**2).sum()

def information_gain(p_A, p_B, l_A, l_B, r_A, r_B):
    return gini(p_A, p_B) - (l_A+l_B)/(p_A+p_B)*gini(left_child_A, left_child_B) - (r_A+r_B)/(p_A+p_B)*gini(r_A, r_B)
```

```python
gini_parent = gini(parent_A, parent_B)
gini_parent
```

```python
gini_child_l = gini(left_child_A, left_child_B)
gini_child_l
```

```python
gini_child_r = gini(right_child_A, right_child_B)
gini_child_r
```

```python
info_gain = information_gain(parent_A, parent_B, left_child_A, left_child_B, right_child_A, right_child_B)
info_gain
```

*혹은*  
- parent = np.array([parent_A, parent_B])  형식으로도 구현할 수 있다!
- gini(parent)
